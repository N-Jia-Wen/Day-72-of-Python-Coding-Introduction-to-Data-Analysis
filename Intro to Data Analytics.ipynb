```python
low_risk = new_data.sort_values('Spread')
```

```python
low_risk[['Undergraduate Major', 'Spread']].head()
```

```python
highest_potential = new_data.sort_values('Mid-Career 90th Percentile Salary', ascending=False)
highest_potential[['Undergraduate Major', 'Mid-Career 90th Percentile Salary']].head()
```

```python
greatest_spread = new_data.sort_values('Spread', ascending=False)
greatest_spread[['Undergraduate Major', 'Spread']].head()
```
